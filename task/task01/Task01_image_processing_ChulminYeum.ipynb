## Image Size

```python
import matplotlib.pyplot as plt
from PIL import Image

# directory of sample images
sampImgDir = 'img'

# image filename
filenameImg = f'{sampImgDir}/car1.jpg'

# read an image
img = Image.open(filenameImg)

# show the image
plt.imshow(img)
plt.axis('off')  # to hide axes
plt.show()

# Obtain image dimensions
w, h = img.size  # This returns (width, height) contrary to MATLAB's (height, width)
c = len(img.getbands())

print(f'width: {w}, height: {h}, and channel(depth): {c}')

# Using PIL to check if the image is RGB or grayscale
if img.mode == 'RGB':
    c = 3
else:
    c = 1
print(f'width: {w}, height: {h}, and channel(depth): {c}')
```

## Grayscale Image

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating grayscale images
grey0 = np.zeros((300, 300), dtype=np.uint8)
grey100 = np.ones((300, 300), dtype=np.uint8) * 100
grey200 = np.ones((300, 300), dtype=np.uint8) * 200
grey255 = np.ones((300, 300), dtype=np.uint8) * 255
grey300 = np.ones((300, 300), dtype=np.uint8) * 300

# Displaying the images
plt.figure(1, figsize=(15, 3))
plt.subplot(151)
plt.imshow(grey0, cmap='gray', vmin=0, vmax=255)
plt.title('Grey 0')
plt.subplot(152)
plt.imshow(grey100, cmap='gray', vmin=0, vmax=255)
plt.title('Grey 100')
plt.subplot(153)
plt.imshow(grey200, cmap='gray', vmin=0, vmax=255)
plt.title('Grey 200')
plt.subplot(154)
plt.imshow(grey255, cmap='gray', vmin=0, vmax=255)
plt.title('Grey 255')
plt.subplot(155)
plt.imshow(grey300, cmap='gray', vmin=0, vmax=255)
plt.title('Grey 300')
plt.show()
```

## Color Image

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os

# Directory of sample images
sampImgDir = 'img'

# Image filename
filenameImgC = f'{sampImgDir}/autumn.tif'

# Reading the first image
imgC = Image.open(filenameImgC)
imgC_np = np.array(imgC)

# Separating RGB channels
imgCR = imgC_np[:, :, 0]
imgCG = imgC_np[:, :, 1]
imgCB = imgC_np[:, :, 2]

# Displaying the red channel and the grayscale image
plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(imgCR, cmap='Reds')
plt.subplot(122)
plt.imshow(imgC.convert('L'), cmap='gray')
plt.show()

# Creating images with individual color channels
imgCRImg = np.stack((imgCR, np.zeros_like(imgCR), np.zeros_like(imgCR)), axis=2)
imgCGImg = np.stack((np.zeros_like(imgCG), imgCG, np.zeros_like(imgCG)), axis=2)
imgCBImg = np.stack((np.zeros_like(imgCB), np.zeros_like(imgCB), imgCB), axis=2)

# Displaying the images with individual color channels
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.imshow(imgCRImg)
plt.subplot(132)
plt.imshow(imgCGImg)
plt.subplot(133)
plt.imshow(imgCBImg)
plt.show()

# Displaying the combined color channels
plt.figure()
plt.imshow(imgCRImg + imgCGImg + imgCBImg)
plt.show()

# Directory of sample images
sampImgDir = 'img'

# Image filename
filenameImg = f'{sampImgDir}/pexels-photo-325044.jpeg'

# Reading the second image
img = Image.open(filenameImg)
img_np = np.array(img)

# Displaying the second image and its individual channels
plt.figure(figsize=(10, 8))
plt.subplot(221)
plt.imshow(img)
plt.subplot(222)
plt.imshow(img_np[:, :, 0], cmap='Reds')
plt.subplot(223)
plt.imshow(img_np[:, :, 1], cmap='Greens')
plt.subplot(224)
plt.imshow(img_np[:, :, 2], cmap='Blues')
plt.show()
```

## Add Circles on the Image

```python
import cv2
import matplotlib.pyplot as plt

# Directory of sample images and image filename
sampImgDir = 'img'
filenameImg = sampImgDir + '/kids.tif'

# Read the image
img = cv2.imread(filenameImg)

# Convert from BGR to RGB
# OpenCV reads images in BGR (Blue, Green, Red) format, while matplotlib displays images in RGB (Red, Green, Blue) format.
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image with two points
plt.figure()

# First subplot with original image and two points marked
plt.subplot(121)
plt.imshow(img)
plt.scatter(50, 150, s=30, facecolors='none', edgecolors='r')  # Red point
plt.scatter(150, 50, s=30, facecolors='none', edgecolors='b')  # Blue point

# Second subplot with circles inserted around the points
img2 = cv2.circle(img.copy(), (50, 150), 5, (255, 0, 0), 1)  # Red circle
img2 = cv2.circle(img2, (150, 50), 5, (0, 0, 255), 1)  # Blue circle

plt.subplot(122)
plt.imshow(img2)
plt.show()
```

## Adjust Image Contrast Values

```python
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt

sampImgDir = 'img'
img_path = sampImgDir + '/pout.tif'

# Open the image using PIL and convert to grayscale for simplicity
img = Image.open(img_path).convert('L')

# Displaying original and adjusted images using PIL's equalize function
img_adjusted = ImageOps.equalize(img)

plt.figure(figsize=(14, 10))
plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.title('Original')

plt.subplot(222)
plt.hist(np.array(img).ravel(), bins=256, range=[0, 256])
plt.title('Original Histogram')


plt.subplot(223)
plt.imshow(img_adjusted, cmap='gray')
plt.title('Adjusted')
plt.subplot(224)
plt.hist(np.array(img_adjusted).ravel(), bins=256, range=[0, 256])
plt.title('Adjusted Histogram')

# Manual contrast adjustment (simple method)
# Here is how we adjust pixel values
# You are going to change each of pixel values following the above rule
# Old value i in the range of 0 ~ indMin => New values: 0
# Old value i in the range of indMin+1 ... indMax-2 => New values: 255/(indMax-indMin-1).*(img-indMin)
# Old value i in the range of indMax-1 ~ 255 => New values: 255

hist, hist_edges = np.histogram(np.array(img).ravel(), bins=256, range=[0, 256])
centers = (hist_edges[:-1] + hist_edges[1:]) / 2

# Calculating the cumulative distribution
B = np.cumsum(hist).astype(float) / hist.sum()

plt.figure()
plt.plot(centers, B, linewidth=3, color='blue')
plt.title('Cumulative Distribution')

# Finding indMin and indMax to rescale the image
indMin = np.max(np.where(B < 0.01))
indMax = np.min(np.where(B >= 0.99))

print(f'New range from {indMin} to {indMax}')

# Contrast adjustment function using PIL
def contrast_adjust_pil(image, indMin, indMax):
    img_np = np.array(image, dtype=float)
    img = (img_np - indMin) * (255.0 / (indMax - indMin - 1.0))
    img[img_np <= indMin] = 0
    img[img_np >= indMax] = 255
    return Image.fromarray(img.astype(np.uint8))

# Applying the manual contrast adjustment
img_manual_adjusted = contrast_adjust_pil(img, indMin, indMax)

# Displaying the manually adjusted image
plt.figure(figsize=(14, 5))
plt.subplot(121)
plt.imshow(img_manual_adjusted, cmap='gray')
plt.title('Original Histogram')

plt.subplot(122)
plt.hist(np.array(img_manual_adjusted).ravel(), bins=256, range=[0, 256])
plt.title('Adjusted Histogram')

plt.show()
```

## Spatial Resolution

```python
from PIL import Image
import matplotlib.pyplot as plt

# Directory of sample images
sampImgDir = 'img'

# Image filename
filenameImg = f'{sampImgDir}/pout.tif'

# Read an image
img = Image.open(filenameImg)

# Resize the image to 10% of its original size
img01 = img.resize((img.size[0]//10, img.size[1]//10), Image.NEAREST)

# Resize the 10% image back to its original size
img10 = img01.resize(img.size, Image.NEAREST)


plt.figure(1, figsize=(13, 3))
plt.subplot(131)
plt.imshow(img)
plt.title(f'{img.size[0]} x {img.size[1]}')

plt.subplot(132)
plt.imshow(img01)
plt.title(f'{img01.size[0]} x {img01.size[1]}')

plt.subplot(133)
plt.imshow(img10)
plt.title(f'{img10.size[0]} x {img10.size[1]}')

plt.show()
```

## Image File Sizes

```python
from PIL import Image
import os
import sys

# Directory of sample images
sampImgDir = 'img'

# Image filenames
filenameImgC = os.path.join(sampImgDir, 'autumn.tif')
filenameImgG = os.path.join(sampImgDir, 'football.jpg')

# Read images
imgC = Image.open(filenameImgC)
imgG = Image.open(filenameImgG)

# Function to calculate image size in bytes
def calculate_image_size(image, filename):
    
    file_stats = os.stat(filename)

    depth_byte = len(image.getbands())
    imgByte = image.size[0] * image.size[1] * depth_byte

    return imgByte, file_stats.st_size

# Calculate sizes and compare
imgCByte, fileCSize = calculate_image_size(imgC, filenameImgC)
imgGByte, fileGSize = calculate_image_size(imgG, filenameImgG)

# Print sizes
print(f'The file size of imgC in bytes: {fileCSize} and its size in bytes after loading into workspace: {imgCByte}')
print(f'The file size of imgG in bytes: {fileGSize} and its size in bytes after loading into workspace: {imgGByte}')
```
